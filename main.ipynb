```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')

df = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Skill-India-AI-ML-Scholarship/main/Summer%20Olympics/summer.csv')

df.head()
```

```python
#1. In how many cities Summer Olympics is held so far?
print("Total no. of Cities :", len(df['City'].unique()))
```

```python
#2. Which sport is having most number of Gold Medals so far? (Top 5)
gold_df = df[df['Medal'] == 'Gold']
data = [[sport, len(gold_df[gold_df['Sport'] == sport])] for sport in gold_df['Sport'].unique()]

data = pd.DataFrame(data, columns=['Sport', 'Medal'])
data.sort_values(by = 'Medal', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Sport'][:5], data['Medal'][:5], color = 'orange', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Sports with most Gold Medals')
plt.ylabel('No of Gold Medals')
plt.xlabel('Sport')
```

```python
#3. Which sport is having most number of medals so far? (Top 5)
data = [[sport, len(df[df['Sport'] == sport])] for sport in df['Sport'].unique()]
data = pd.DataFrame(data, columns=['Sport', 'Medal'])
data.sort_values(by = 'Medal', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Sport'][:5], data['Medal'][:5], color = 'aqua', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Sports with most no. of Medals')
plt.ylabel('No of Medals')
plt.xlabel('Sport')
```

```python
#4. Which player has won most number of medals? (Top 5)
data = [[Athlete, len(df[df['Athlete'] == Athlete])] for Athlete in df['Athlete'].unique()]
data = pd.DataFrame(data, columns=['Athlete', 'Medal'])
data.sort_values(by = 'Medal', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Athlete'][:5], data['Medal'][:5], color = 'lightgreen', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Player with most no. of Medals')
plt.ylabel('No of Medals')
plt.xlabel('Player Name')
```

```python
#5. Which player has won most number Gold Medals of medals? (Top 5)
gold_df = df[df['Medal'] == 'Gold']
data = [[Athlete, len(gold_df[gold_df['Athlete'] == Athlete])] for Athlete in gold_df['Athlete'].unique()]
data = pd.DataFrame(data, columns=['Athlete', 'Medal'])
data.sort_values(by = 'Medal', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Athlete'][:5], data['Medal'][:5], color = 'limegreen', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Player with most no. of Gold Medals')
plt.ylabel('No of Gold Medals')
plt.xlabel('Player Name')
```

```python
#6. In which year India won first Gold Medal in Summer Olympics?
india_df = df[df['Country'] == 'IND'].sort_values(by= 'Year')
print("India won First Gold Medal in", int((india_df[india_df['Medal'] == 'Gold'])['Year'][0:1]))
```

```python
#7. Which event is most popular in terms on number of players? (Top 5)
data = [[Event, len(df[df['Event'] == Event])] for Event in df['Event'].unique()]
data = pd.DataFrame(data, columns=['Event', 'No of Participation'])
data.sort_values(by = 'No of Participation', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Event'][:5], data['No of Participation'][:5], color = 'pink', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Most popular events')
plt.ylabel('No of Players participated')
plt.xlabel('Event Name')
```

```python
#8. Which sport is having most female Gold Medalists? (Top 5)
female_df = df[df['Gender'] == 'Women']
gold_df = female_df[female_df['Medal'] == 'Gold']
data = [[Sport, len(gold_df[gold_df['Sport'] == Sport])] for Sport in gold_df['Sport'].unique()]
data = pd.DataFrame(data, columns=['Sport', 'No of Female players'])
data.sort_values(by = 'No of Female players', ascending = False, inplace = True)

plt.figure(figsize = (20,10))
barplot = plt.bar(data['Sport'][:5], data['No of Female players'][:5], color = 'red', alpha = .8)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.title('Sport having most no. of Female Gold Medalists')
plt.ylabel('No of Female players')
plt.xlabel('Sport Name')
```
